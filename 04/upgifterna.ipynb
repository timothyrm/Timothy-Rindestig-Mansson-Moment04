```python
#3
from math import sqrt

counter = 0
s1 = []
s2 = []
arealist = []
höjd = []
volymlist = []
kvadratlist = []

def area(y,x):
    a = y*x
    return a

def volym(z,y,x):
    v = z*y*x
    return v

def kvadrat(x,y):
    if x == y: 
        return "ja"


while True:
    svar = input("Vill du fortsätta göra beräkningar? J / N?").upper()
    if svar == "J":
        counter +=1

        y=int(input("ange ett heltal för sida a"))
        if y < 0:
            y = 1  
        s1.append(y)
   
        x=int(input("ange ett heltal för sida b"))
        if x < 0:
            x = 1  
        s2.append(x)
    
        z=int(input("ange en höjd för rektanglen"))
        if z < 0:
            z = 1  
        höjd.append(z)


        arealist.append(area(x,y))
        volymlist.append(volym(z,y,x))
        kvadratlist.append(kvadrat(x,y))

        if kvadrat(x,y) == "ja":
            print(f"arean på rektangel med sidorna {y} och {x} är {area(y,x)} och det är en kvadrat.")
        else:
            print(f"arean på rektangel med sidorna {y} och {x} är {area(y,x)}")
        

        print("Höjd | Volym")
        print("------------")
        for c in range(1,11):
            print(c,"|",y*x*c)
        
    else: 
        for i in range(counter):
            if kvadratlist[i] == "ja":
                print(f"Rektangeln har sidorna {s1[i]} och {s2[i]} vilket gör att arean är {arealist[i]}, rektangeln är en kvadrat.")
            else:
                print(f"Rektangeln har sidorna {s1[i]} och {s2[i]} vilket gör att arean är {arealist[i]}")            
        break
```
